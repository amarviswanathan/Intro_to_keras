```python
concrete_data = pd.read_csv('https://ibm.box.com/shared/static/svl8tu7cmod6tizo6rk0ke4sbuhtpdfx.csv')
concrete_data.head()
```

```python
concrete_data_columns = concrete_data.columns
concrete_data_filtered = concrete_data_columns[concrete_data_columns != 'Strength']
X = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']]
y = concrete_data['Strength']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X[concrete_data_filtered] = scaler.fit_transform(X[concrete_data_filtered])
X.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense
def regression_model():
    
    model = Sequential()
    model.add(Dense(50,activation='relu',input_shape = (X.shape[1],)))
    model.add(Dense(50,activation='relu'))
    model.add(Dense(1))
    
    model.compile(optimizer='adam',loss='mean_squared_error')
    return model
```

```python
model = regression_model()
model.fit(X,y,validation_split=0.3,epochs=100,verbose=2)
```
